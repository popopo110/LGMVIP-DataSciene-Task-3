```python
# import pandas lib as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
# read by default 1st sheet of an excel file
df = pd.read_excel('globalterrorism.xlsx')

df.head()
```

```python
df
```

```python
#Adjust the presented 
df=df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region',
                             'provstate':'State','city':'City','attacktype1_txt':'Attack_type','targtype1_txt':'Target_type',
                             'targsubtype1_txt':'Target_subtype','gname':'Gang_name','weaptype1_txt':'Weapon_type'})
```

```python
df
```

```python
df=df[['Year','Month','Day','Country','Region','State','City','Attack_type','Target_type','Target_subtype','Gang_name','Weapon_type']]
```

```python
df
```

```python
df['Target_subtype'] = df['Target_subtype'].fillna('others')
```

```python
df['City']=df['City'].fillna('Unknown')
```

```python
df['State']=df['State'].fillna('Unknown')
```

```python
df.replace( {"Vehicle (not to include vehicle-borne explosives, i.e., car or truck bombs)": "Vehicle"},inplace = True)
```

```python
df
```

```python
#check shape
df.shape
```

# Information about each category

```python
def total():
    return df['Country'].nunique(),df['Country'].unique()
def Target():
    return df['Target_type'].unique(),df['Target_type'].nunique()
def weapon(): 
    return df['Weapon_type'].unique(),df['Weapon_type'].nunique()
def attack():
    return df['Attack_type'].unique(), df['Attack_type'].nunique()
def region():
    return df['Region'].unique(), df['Region'].nunique()
```
